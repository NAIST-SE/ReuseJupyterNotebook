```python
import os
import csv
import codecs
import json
from datetime import datetime

abs_path = '/mnt/ext-intern/2022-intern-data/share-intern-data_2022/kgtorrent_2022/Palms_data/experiment/RQ1'
os.chdir(abs_path)

src_dir = 'competitors'

competitor_list = []
file_list = os.listdir(src_dir)

for file in file_list:
    competitor_id = file[:-4]
    if file[-4:] == '.csv':
        
        if os.path.exists(competitor_id):
            os.system('rm -rf ' + competitor_id)
        # end if
        
        if not os.path.exists(competitor_id):
            os.system('mkdir ' + competitor_id)
            competitor_list.append(competitor_id)
        # end if
    # end if
# end for
```

```python
file_counter = {}
empty_competitor_folder = []

competitors_metadata = {}

for competitor in competitor_list:
    
    kernel_date = {}
    count = 0
    total_count = 0
    with open(src_dir + '/' + competitor + '.csv', 'r') as metadata:
        lines = csv.reader(metadata)
        next(lines, None)  # skip the headers
        
        for line in lines:
            path = './' + competitor + '/' + line[1]
            
            if not os.path.exists(path):
                os.system('mkdir ' + path)
            # end if
                
            file_path = '../../../KT_dataset/' + line[0]
            comp_id = line[1]
            kernel_date[comp_id] = datetime.strptime(line[2], '%Y-%m-%d %H:%M:%S')
            total_count = total_count + 1

            if os.path.exists(file_path):
                command = 'cp ' + file_path + ' ./' + competitor + '/' + comp_id
                os.system(command)
                count = count + 1
            # end if
        # end for
    # end with
    
    file_counter[competitor] = [count, total_count]
    if count == 0: empty_competitor_folder.append(competitor)
        
    # sort comp_id by kernel_created_date
    kernel_date = dict(sorted(kernel_date.items(), key=lambda item: item[1]))
    competitors_metadata[competitor] = kernel_date
# end for

for each_competitor in file_counter:
    not_found = file_counter[each_competitor][1] - file_counter[each_competitor][0]
    print(
        'Copied %(counter)d/%(total_count)d file(s) -> Competitor: %(comp)s' 
        % {
            'counter': file_counter[each_competitor][0], 
            'total_count':file_counter[each_competitor][1], 
            'comp': each_competitor
          }
    )
    if not_found != 0:
        print('[%d file(s) not found]\n' % not_found)
    else:
        print('*[All files are found]\n')
    # end if
# end for
```

```python
file_list = os.listdir('.')
all_empty_dir = {}

for file in file_list:
    if file[0] == '.' or file == src_dir: continue
    
    file_counter = 0
    
    if os.path.isdir(file):
        competitor_dir = os.listdir(file)
        empty_dir = []
        print('[%s]' % file)
        
        for comp_id in competitor_dir:
            comp_file_lists = os.listdir(file + '/' + comp_id)
            current_file_path = file + '/' + comp_id
            index = 0
            for each_comp_file in comp_file_lists:
                if each_comp_file[-6:] == '.ipynb':
                    notebook_file_name = current_file_path + '/' + each_comp_file

                    f = codecs.open(notebook_file_name, 'r')
                    source = f.read()
                    notebook = json.loads(source)
                    
                    for row in notebook['cells']:
                        if row['cell_type'] == 'code':
                            line_count = 0
                            source_length = len(row['source'])
                            new_file_path = current_file_path + '/' + str(index) + '.py'

                            with open(new_file_path, 'w+') as new_file:
                                for line in row['source']:
                                    new_file.write(line)
                                    if line_count == (source_length - 1):
                                        if line[-1] != '\n': new_file.write('\n')
                                    # end if
                                    line_count = line_count + 1
                                # end for
                            # end with
                            index = index + 1
                        # end if
                    # end for
                # end if
                if os.path.exists(notebook_file_name): os.system('rm ' + notebook_file_name)
            # end for
            if index == 0: empty_dir.append(comp_id)
            print('%(i)d files are created -> %(p)s' % {'i': index, 'p': current_file_path}) 
        # end for
        all_empty_dir[file] = empty_dir
        print('')
    # end if 
# end for
```

```python
# competition_amount = 10
only_ten_comp = {}

for competitor in competitor_list:
    empty_dir = all_empty_dir[competitor]
    comp_with_time = {}
    for key in competitors_metadata[competitor]:
        if key not in empty_dir:
            comp_with_time[key] = competitors_metadata[competitor][key]
        # end if
    # end for
    comp_with_time = dict(sorted(comp_with_time.items(), key=lambda item: item[1]))
    only_ten_comp[competitor] = comp_with_time
# end for
```

```python
for each in empty_competitor_folder:
    if os.path.exists(each):
        os.system('rm -rf ' + each)
        print('Removed: ./%s directory' % each)
    # end if
# end for

competition_counter = {}
file_list = os.listdir('.')

for file in file_list:
    if os.path.isdir(file) and file[0] != '.' and file != src_dir:
        counter = 0
        competitor_dir = os.listdir('./' + file)
        for comp_id in competitor_dir:
            cur_path = './' + file + '/' + comp_id
            if comp_id in only_ten_comp[file]:
                competition_dir = os.listdir(cur_path)
                if len(competition_dir) == 0:
                    os.system('rm -rf ' + cur_path)
                    print('[Removed: %s]' % cur_path)
                else:
                    counter = counter + 1
                # end if
            else:
                os.system('rm -rf ' + cur_path)
                print('[Removed: %s]' % cur_path)
            # end if
        # end for
        competition_counter[file] = counter
    # end if
# end for
```
